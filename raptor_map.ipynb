Import requires packages

```python
import json
import pandas as pd
import geopandas as gpd
from keplergl import KeplerGl
```

Create GeoDataFrame of RSPB Raptor Incidents

```python
gdf = gpd.read_file('raptor_incidents.geojson')
gdf.head()
```

Filter on find incidents involving Red Kites

```python
rk_gdf = gdf[gdf['Details'].str.contains("red kite", case=False)]
rk_gdf.head()
```

Create lat/long from geometry centroid so that we can create a heatmap in Kepler

```python
rk_gdf['Latitude'] = rk_gdf.geometry.centroid.y
rk_gdf['Longitude'] = rk_gdf.geometry.centroid.x
rk_gdf.head()
```

Set up map. Data is added twice - once for the heatmap, once for the tooltip to display. Save map to HTML for deployment.

```python
rk_points = pd.DataFrame(rk_gdf.drop("geometry", axis=1))
rk_heatmap = pd.DataFrame(rk_gdf.drop("geometry", axis=1))
rk_points.to_csv('rk_points.csv', index=False)
rk_heatmap.to_csv('rk_heatmap.csv', index=False)
config = json.load(open('config.json'))
map = KeplerGl(config=config)
map.add_data(rk_points, name="Incidents")
map.add_data(rk_heatmap, name="Incidents - Heat Map")
map.save_to_html(file_name="index.html")
map
```
